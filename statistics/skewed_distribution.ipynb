```python
%matplotlib inline
import scipy.stats as stats
import ipywidgets as widgets
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
```

# Skewed Normal Distribution

This section gives an introduction to the skewed normal distribution, it attempts to give a feel for the distribution.

```python
x = np.linspace(-5,5,100)

def plot(alpha: float, loc: float, scale: float) -> None:
    fix, ax = plt.subplots(figsize=(10,5))
    pdf = stats.skewnorm.pdf(x=x, a=alpha, loc=loc, scale=scale)
    ax.plot(x, pdf, color='red', label='Skew')
    ax.set_ylim(0, 1.0)
    plt.show()



interactive_plot = widgets.interactive(plot, alpha=(-5.0, 5.0), loc=(-2.0, 2.0), scale=(0.5, 3.0))
output = interactive_plot.children[-1]
output.layout.height = '100%'
interactive_plot
```
