# Skewed Normal Distribution

The skewed normal distribution is given by


$f(x) = \frac{2}{\sigma}\phi\left(\frac{x-\mu}{\sigma}\right)\Phi\left(\alpha\left(\frac{x-\mu}{\sigma}\right)\right)$

where $\phi$ is the standard normal pdf and $\Phi$ is the standard normal cdf. The parameters are:
- $\mu$ is the location parameter
- $\sigma$ is the scale parameter
- $\alpha$ is the shape parameter. If $\alpha=0$, the distribution is normal


```python
x = np.linspace(-5,5,100)

options = [
    (-3.0, 0, 1.0),
    (5.0, 0, 1.0),
    (5.0, 2.0, 1.0),
    (5.0, -2.0, 1.0),
    (5.0, -2.0, .5),
    (5.0, -2.0, 3.0),
]

fix, axes = plt.subplots(
    len(options)//2, 2,
    figsize=(20, 10),
    sharex=True,
    sharey=True,
    constrained_layout=True,
)

for no, (alpha, loc, scale) in enumerate(options):
    pdf = stats.skewnorm.pdf(x=x, a=alpha, loc=loc, scale=scale)
    ax = axes[no//2, no%2]
    ax.plot(x, pdf, color='black')
    ax.vlines(loc, 0, 1.5)
    ax.set_ylim(0, 1.5)
    ax.set_xlim(x.min(), x.max())
    title = f"$\\alpha={alpha}, \\mu={loc}, \\sigma={scale}$"
    ax.set_title(title, color='black', fontsize=20)
```
